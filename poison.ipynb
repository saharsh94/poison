```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np



datapath= '/content/drive/MyDrive/Colab Notebooks/sam/41598_2018_24483_MOESM3_ESM (1).xlsx'
Data=pd.read_excel(datapath,sheet_name='Sheet1')
#Data = Data.astype(np.float)

label = np.array(Data['Target'])


#print(DataF.columns)
a = np.delete(Data.to_numpy(),19,axis=1)
a = np.delete(a,14,axis=1)
print(a)
print(a.shape)
print(label)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

train_data, test_data, train_labels, test_labels = train_test_split(Data, label, test_size=0.3, random_state=20)

clf = DecisionTreeClassifier(random_state=42)
clf.fit(train_data, train_labels)
predictions = clf.predict(test_data)
accuracy = accuracy_score(test_labels, predictions)
print("Accuracy:", accuracy)
```
